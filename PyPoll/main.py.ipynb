```python
# -*- coding: UTF-8 -*-
"""PyPoll Homework Starter File."""

# Import necessary modules
import csv
import os

# Files to load and output (update with correct file paths)
file_to_load = os.path.join("Resources", "election_data.csv")  # Input file path
file_to_output = os.path.join("analysis", "election_analysis.txt")  # Output file path

ballot_ID = []
cand_list = [] #top2bottom candidate list
candidates = [] #unique candidate names
vote_count = [] #holds each candidate count >> vote_count = {}?
percent_vote = []


with open(file_to_load) as election_data:
    reader = csv.reader(election_data)
    header = next(reader)

# Loop through each row of the dataset and process it
    # Increment the total vote count for each row
           # If the candidate is not already in the candidate list, add them
       #Add a vote to the candidate's count

    for row in reader:
        ballot_ID.append(row[0])
        total_votes = len(ballot_ID)
    
        cand_list.append(row[2])
        
# Get the candidate's name from the row        
    # realized I couldn't index set(list)
    for each_cand in set(cand_list):       
        candidates.append(each_cand)
        
# votes per candidate
        votes = cand_list.count(each_cand)
        vote_count.append(votes)

# percent votes per candidate
        percent = round((votes/total_votes)*100,3)
        percent_vote.append(percent)

    max_vote = max(vote_count)
    winner = candidates[vote_count.index(max_vote)]
    
# Print the total vote count (to terminal)

print("Election Results")   
print("-------------------------")
print(f"Total Votes: {total_votes}")   
print("-------------------------")
for c in range(len(candidates)): print(f"{candidates[c]}: {percent_vote[c]}%  ({vote_count[c]})")
print("-------------------------")
print(f"Winner: {winner}")
print("-------------------------")

# Open a text file to save the output
with open(file_to_output, "w") as PyPoll_txt:

    PyPoll_txt.write("Election Results\n")
    PyPoll_txt.write("-------------------------\n")
    PyPoll_txt.write(f"Total Votes: {total_votes}\n")   
    PyPoll_txt.write("-------------------------\n")
    for c in range(len(candidates)): PyPoll_txt.write(f"{candidates[c]}: {percent_vote[c]}%  ({vote_count[c]})\n")
    PyPoll_txt.write("-------------------------\n")
    PyPoll_txt.write(f"Winner: {winner}\n")
    PyPoll_txt.write("-------------------------")
```
